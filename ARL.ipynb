## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# !pip install apyori
from apyori import apriori
```

```python
data = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
data.shape
```

```python
data.head()
```

```python
transactions = []
for i in range(7500):
    transactions.append([str(data.values[i, j]) for j in range(20)])
pd.DataFrame(transactions).head()
```

```python
## fitting the apriori model
rules = apriori(transactions=transactions, 
                min_support = 0.003, 
                min_confidence = 0.2, 
                min_lift = 3, 
                min_length = 2, 
                max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, support, confidence, lifts))
df = pd.DataFrame(inspect(results), columns = ['If buy', 'Also Buy', 'Support', 'Confidence', 'Lift'])
df
```
